```python
#check columns have invalid valus

# Define validation rules
def is_valid_star_rating(x):
    numeric_x = pd.to_numeric(x, errors='coerce')
    return numeric_x.between(0, 5)

def is_valid_rating(x):
    numeric_x = pd.to_numeric(x, errors='coerce')
    return numeric_x.between(0.0, 10.0)

def is_valid_yes_no(x):
    return x.isin(['Yes', 'No'])


validation_rules = {
    'Star Rating': is_valid_star_rating,
    'Rating': is_valid_rating,
    'Free Parking': is_valid_yes_no,
    'Fitness Centre': is_valid_yes_no,
    'Spa and Wellness Centre': is_valid_yes_no,
    'Airport Shuttle': is_valid_yes_no,
    'Staff': is_valid_rating,
    'Facilities': is_valid_rating,
    'Location': is_valid_rating,
    'Comfort': is_valid_rating,
    'Cleanliness': is_valid_rating
}

# Function to check invalid values for each column based on defined rules
def check_invalid_values(data, rules):
    invalid_counts = {}
    for column, rule in rules.items():
        if column in data.columns:
            invalid_mask = ~rule(data[column])
            invalid_count = invalid_mask.sum()
            invalid_counts[column] = invalid_count
            print(f"Number of invalid values in {column} -----> {invalid_count}")

    return invalid_counts



# Check invalid values
invalid_counts = check_invalid_values(data, validation_rules)
```

```python
# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(list(invalid_counts.keys()), list(invalid_counts.values()), marker='o',color ='green')
plt.title('Number of Invalid Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Number of Invalid Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Model Building

```python
#select features in x and y
x = data.drop('Price Per Day($)',axis=1)
y = data['Price Per Day($)']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Display the shapes of the resulting datasets
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

In this problem is Regression problem.So we use Linear Regression Algorithem.

```python
#import necessary algorithems
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics

# loading the linear regression model
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)

# Making predictions
y_train_pred = lin_reg_model.predict(x_train)
y_test_pred = lin_reg_model.predict(x_test)


# Calculating Mean Squared Error (MSE)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

# Calculating R-squared (RÂ²) for accuracy
train_r2 = metrics.r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)


# Displaying the results
print(f"Training MSE: {train_mse}")
print(f"Test MSE: {test_mse}")
print(f"Training RÂ²: {train_r2}")
print(f"Test RÂ²: {test_r2}")
```

```python
# The line of best fit for training data
m, b = np.polyfit(y_train, y_train_pred, 1)
plt.plot(y_train, m*y_train + b, color='red', label='Regression Line (Train)')

plt.scatter(y_train, y_train_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# The line of best fit for training data
m, b = np.polyfit(y_test, y_test_pred, 1)
plt.plot(y_train, m*y_train + b, color='red', label='Regression Line (Train)')

plt.scatter(y_test, y_test_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
